```python
import pandas as pd
import numpy as np
```

```python
allocations = pd.read_csv("/home/sagar/Dropbox/Scheduler/reports/21Nov/binary_class.csv")
```

```python
allocations.set_index(['Date'],drop=True,inplace=True)
```

```python
allocations.head()
```

```python
allocations = allocations.applymap(lambda x: 1 if x>0 else 0)
```

```python
allocations.fillna(0,inplace=True)
```

```python
allocations['Total Classes'] = allocations.apply(np.sum, axis=1)
```

```python
allocations['Total DS'] = allocations['Total Classes'].apply(lambda x: x if x==2 else 3*x) 
```
